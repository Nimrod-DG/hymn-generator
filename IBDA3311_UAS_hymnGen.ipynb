### Fine-tuning

Lakukan fine-tuning (penyetelan lebih lanjut) model GPT-2 pada dataset spesifik dengan menggunakan pustaka gpt_2_simple.

1. **file_name = "data/indonesian_lyrics.txt"**: Ini mendefinisikan nama file dan path ke dataset yang akan digunakan untuk fine-tuning. Dalam hal ini, dataset berupa teks lirik bahasa Indonesia, dan file tersebut disimpan di dalam direktori "data" dengan nama file "indonesian_lyrics.txt".

2. **sess = gpt2.start_tf_sess()**: Ini memulai sesi TensorFlow yang akan digunakan untuk fine-tuning model. Sesuai dengan konvensi pustaka gpt_2_simple, sesi TensorFlow harus dimulai sebelum menggunakan fungsi-fungsi fine-tuning.

3. **gpt2.finetune(...)**: Ini adalah pemanggilan fungsi untuk melakukan fine-tuning pada model GPT-2. Parameter yang digunakan adalah sebagai berikut:

        sess: Parameter ini adalah sesi TensorFlow yang sudah dimulai sebelumnya.

        dataset=file_name: Parameter ini menentukan dataset yang akan digunakan untuk fine-tuning. Dalam hal ini, menggunakan dataset yang telah didefinisikan sebelumnya.

        model_name='124M': Parameter ini menentukan versi model GPT-2 yang akan digunakan. Dalam hal ini, menggunakan model dengan 124 juta parameter.

        steps=2600: Parameter ini menentukan jumlah langkah fine-tuning yang akan dijalankan.

        restore_from='latest': Parameter ini menentukan dari mana model akan dipulihkan sebelum memulai fine-tuning. 'latest' berarti menggunakan model terakhir yang telah diunduh atau fine-tuned.

        run_name='indo': Parameter ini menentukan nama run untuk hasil fine-tuning. Hasil fine-tuning akan disimpan dalam direktori yang sesuai.

        print_every=100: Parameter ini menentukan seberapa sering akan mencetak informasi tentang proses fine-tuning (setiap 100 langkah).

        sample_every=200: Parameter ini menentukan seberapa sering akan menghasilkan contoh teks menggunakan model fine-tuned (setiap 200 langkah).

        save_every=200: Parameter ini menentukan seberapa sering akan menyimpan model yang telah fine-tuned (setiap 200 langkah).

```python
file_name = "data/indonesian_lyrics.txt"
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='124M',
              steps=2600,
              restore_from='latest',
              run_name='indo',
              print_every=100,
              sample_every=200,
              save_every=200,
              )
```

### Generate

Hasilkan lirik himne dengan menggunakan model GPT-2 yang telah dilatih dan di fine-tune sebelumnya.

1. **sess = gpt2.start_tf_sess()**: Ini memulai sesi TensorFlow yang akan digunakan untuk meng-load model GPT-2 dan menghasilkan teks.

2. **gpt2.load_gpt2(sess, run_name='engl')**: Ini memuat model GPT-2 yang telah dilatih atau disesuaikan sebelumnya. Parameter run_name menentukan nama run yang digunakan saat menyimpan model. Dalam hal ini, menggunakan nama run 'engl'.

3. **gpt2.generate(...)**: Ini adalah pemanggilan fungsi untuk menghasilkan teks menggunakan model GPT-2 yang telah dimuat. Parameter yang digunakan adalah sebagai berikut:

        length=300: Parameter ini menentukan panjang teks yang akan dihasilkan (dalam jumlah token).

        temperature=1: Parameter ini mengendalikan sejauh mana hasil yang dihasilkan bersifat acak. Nilai yang lebih tinggi (misalnya, di atas 1) akan menghasilkan teks yang lebih kreatif dan tidak terduga, sedangkan nilai yang lebih rendah akan membuat teks lebih deterministik.

        prefix="God is good": Parameter ini memberikan awalan teks yang akan dijadikan dasar untuk generasi selanjutnya. Dalam hal ini, teks dimulai dengan frasa "God is good".

        nsamples=1: Parameter ini menentukan berapa banyak sampel teks yang akan dihasilkan.

        batch_size=1: Parameter ini menentukan ukuran batch untuk penghasilan teks. Dalam hal ini, digunakan batch size 1.

        top_k=40: Parameter ini mengendalikan berapa banyak token teratas yang dipertimbangkan dalam penghasilan teks. Nilai yang lebih tinggi akan membuat teks lebih bervariasi.

        top_p=0.9: Parameter ini mengendalikan jumlah token yang dipertimbangkan berdasarkan probabilitas kumulatif. Nilai yang lebih rendah akan membuat teks lebih bervariasi.

```python
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='engl')
gpt2.generate(sess,
            length=300,
            temperature=1,
            prefix="God is good",
            nsamples=1,
            batch_size=1,
            top_k=40,
            top_p=0.9,
            )
```
